```python
import pandas as pd
```

```python
# reading and showing dataframe
df_exam = pd.read_csv('StudentsPerformance.csv')
df_exam
```

## 1. Create Index

```python
import numpy as np
import random
```

```python
# creating non repetitive values for the index
new_index = np.arange(0, 1000)
```

```python
# shuffling the indexes
random.shuffle(new_index)
```

```python
# create a new column with new index
df_exam['new_index'] = new_index
df_exam
```

## 2. Set Index

```python
# set new_index column as index
df_exam.set_index(new_index, inplace=True)
df_exam
```

## Sort Index

```python
# sort dataframe by new index
df_exam.sort_index()
```

```python
# sort descending by index
df_exam.sort_index(ascending=False, inplace=True)
```
