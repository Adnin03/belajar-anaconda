### 1.2 Adding a new column with an array

The dataframe has 100 rows, so we need to create an array with 1000 elements

```python
# import numpy
import numpy as np
```

```python
# create an array of 1000 elements
language_score = np.arange(0,1000)
```

```python
# length of the array
len(language_score)
```

```python
# adding a new column to dataframe with an array
df_exam['language score'] = language_score
df_exam
```

```python
# create random integer numbers between 1 to 100
random_language_score = np.random.randint(25, 100, size=1000)
```

```python
# min value inclusive and high value exclusive
min(random_language_score)
max(random_language_score)
```

```python
# adding a new column to dataframe with an array
df_exam['language score'] = random_language_score
df_exam
```
