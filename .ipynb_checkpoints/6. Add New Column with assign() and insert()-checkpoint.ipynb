## 1. Add in a new column

### 1.1 assign()

When to use assign?
* Add multiple columns in a single line of code
* When you need to overwrite the values of an existing columns (best practice)

It returns a new object (a copy) with all the original columns in addition to the new ones

```python
import numpy as np
```

```python
# create random numbers for our new score columns
score1 = np.random.randint(1, 100, size=1000)
score2 = np.random.randint(1, 100, size=1000)
```

```python
# create a series using the random numbers
series1 = pd.Series(data=score1, index=np.arange(0, 1000))
series2 = pd.Series(data=score2, index=np.arange(0, 1000))
```

```python
series1
```

```python
# using assign() to add multiple columns >> Jadi, score1 sama score2 itu hanya nama saja pada bagian cell ini
df_exam = df_exam.assign(score1 = series1, score2 = series2)
```

```python
# show dataframe
df_exam
```

### 1.2 insert()

Insert a new column at a specifix position or index

```python
df_exam.insert(1, 'test', series1)
```
