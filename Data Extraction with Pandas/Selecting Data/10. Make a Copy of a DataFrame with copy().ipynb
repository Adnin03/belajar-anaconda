```python
import pandas as pd
```

```python
# read csv file
df = pd.read_csv('players_20.csv')
# set index
df.set_index('short_name', inplace=True)
# select column
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality',	'club']]
```

```python
# show dataframe
df
```

## 1. copy()

### 1.1 deep = True

```python
# deep=True by default (Modifications to the data or indices of the copy)
df_copy = df.copy()
```

```python
# update value in original dataframe
df.loc['L. Messi', 'nationality'] = 'Indonesia' 
```

```python
# copy vs original dataframe
df # dataframe asli dan ketika dijalankan L. Messi nationality nya Indonesia karena sudah dirubah pada cell sebelumnya
df_copy # dataframe copy akan menjalankan sesuai dataframe asli sebelum ada perubahan
```

## 1.2 deep=False 

```python
# deep=False (Any changes to the data of the original will be reflected in the shallow copy and vice versa)
df_shallow_copy = df.copy(deep=False)
```

```python
# update value in original dataframe
df.loc['Cristiano Ronaldo', 'height_cm'] = 200
```

```python
# shallow copy vs original dataframe
df
df_shallow_copy
```

### 1.3 Simple assignment

```python
# make a copy
df_new_copy = df
```

```python
# update value in original dataframe
df.loc['Neymar Jr', 'weight_kg'] = 80
```

```python
# copy vs original dataframe
df
df_new_copy
```
