```python
import pandas as pd
```

```python
# read the csv file
df_exam = pd.read_csv('StudentsPerformance.csv')
df_exam.head()
```

## 1. Rename Column 

```python
# rename column and overwrite values of the dataframe
df_exam.rename(columns={'gender':"Gender"}, inplace=True)
```

```python
# rename multiple columns and update the dataframe with inplace argument
df_exam.rename(columns={'math score':'MS', 'reading score':'RS', 'writing score':'WS'}, inplace=True)
```

```python
# show dataframe
df_exam
```

## 2. Rename Index

```python
# rename index 0, 1, and 2 and update the dataframe
df_exam.rename(index={0:'A', 1:'B', 2:'C'}, inplace=True)
```
