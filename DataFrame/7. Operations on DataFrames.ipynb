```python
import pandas as pd
```

```python
# reading the csv file
df_exam = pd.read_csv('StudentsPerformance.csv')
df_exam.head()
```

## 1. Math Operations

### 1.1 Operations in columns

```python
# select a column and calculate total sum
df_exam['reading score'].sum()
```

```python
# count, mean, std, max, and min
df_exam['reading score'].count()
df_exam['reading score'].mean()
df_exam['reading score'].std()
df_exam['reading score'].max()
df_exam['reading score'].min()
```

```python
# easier calculation with .describe()
df_exam.describe()
```

### 1.2 Operations in rows

```python
# Calculating the sum in a row
df_exam['math score'] + df_exam['reading score'] + df_exam['writing score']
```

```python
# Calculating the average score and assigning the result to a new column
df_exam['average'] = (df_exam['math score'] + df_exam['writing score'] + df_exam['reading score'])/3
```

```python
# Showing the dataframe
df_exam.round(2)
```
