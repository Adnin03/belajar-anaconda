```python
import pandas as pd
```

```python
# reading and showing dataframe
df_exam = pd.read_csv('StudentsPerformance.csv')
df_exam
```

## 1. Sort a DataFrame

```python
# sort by one column
df_exam.sort_values('math score')
```

```python
# sort descending by one column 
df_exam.sort_values('math score', ascending=False)
```

```python
# sort descending by multiple columns
df_exam.sort_values(['math score', 'reading score', 'writing score'], ascending=False)
```

```python
# ini saya memanggil dataframe kembali dan hasilnya belum update, padahal sudah saya sort pada cell diatasnya
df_exam
```

```python
# sort descending by multiple columns and update dataframe
df_exam.sort_values(['math score', 'reading score', 'writing score'], ascending=False, inplace=True)
```

```python
# ini adalah pemanggilan dataframe yang sudah di update karena menggunakan atribut inplace
df_exam
```

```python
# sort descending with a key function
df_exam.sort_values('race/ethnicity', key=lambda col:col.str.lower())
```
