## 1. Selecting two or more columns

```python
# Select two columns using [[]]
df_exam[['race/ethnicity', 'math score']]
```

```python
# Select 2 or more columns using [[]]
df_exam[['race/ethnicity', 'reading score', 'writing score', 'math score']]
```
